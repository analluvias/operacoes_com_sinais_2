```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

plt.rcParams['lines.linewidth'] = 2 #largura das linhas
plt.rcParams['axes.grid'] = True #grade
plt.rcParams['font.size']= 13.0 #tamanho da fonte das legendas
plt.rcParams['xtick.labelsize'] = 12 #tamanho da fonte dos valores do eixo x
plt.rcParams['ytick.labelsize'] = 12 #tamanho da fonte dos valores do eixo y
plt.style.use('fivethirtyeight') #estilo
```

```python
def x(t):

  retorno = []

  for i in range(len(t)):
    if(t[i]<-2):
      retorno.append(0)
    elif(t[i]>4):
      retorno.append(0)
    else:
      retorno.append(t[i]**2)

  return retorno

t = np.linspace(-5,5,500)

y1 = x(t)
a=2

plt.plot(t, y1, label='x(t)')
plt.plot(t, a*np.array(y1), label='x(t)')

```

```python
n = np.linspace(-5,5,20)
y2 = x(n)

plt.stem( n, y2, label='x(n)')
plt.stem( n, 2*np.array(y2), label='x(n)')
plt.legend()
```

```python
def x(t):

  retorno = []

  for i in range(len(t)):
    if(t[i]<3):
      retorno.append(0)
    else:
      retorno.append(t[i]**2)

  return retorno

t = np.linspace(-5,10,500)
y = x(t)

plt.plot(t, y)
```

```python
# x(1 - t)

y = x(1-t)
plt.plot(t, y)
```

```python
# x(1 - t) + x(2-t)

y = np.add(x(1-t),x(2-t))
plt.plot(t, y)
```

```python
# x(1 - t) + x(2-t)

y = np.multiply(x(1-t),x(2-t))
plt.plot(t, y)
```

```python
# x(3t)

y = x(3*t)
plt.plot(t, y)
```

```python
# x(t/3)

y = x(t/3)
plt.plot(t, y)
```
